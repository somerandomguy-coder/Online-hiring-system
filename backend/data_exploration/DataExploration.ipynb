```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
import re
def clean_text(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
  text = re.sub(r'\s+', ' ', text)
  text = " ".join(word for word in text.split() if word not in stopwords.words('english'))
  return text
cleaned_sample = clean_text(sample)
print(sample)
print("-"*10)

print(cleaned_sample)
```

```python
df1["cleaned_text"] = df1["Resume_str"].apply(clean_text)
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=30000)
X = vectorizer.fit_transform(df1["cleaned_text"])
```

```python
print(len(vectorizer.get_feature_names_out()))
```

```python
from sklearn.utils import sparsefuncs
means, variances = sparsefuncs.mean_variance_axis(X, axis=0)
means_resume = means
print(means_resume)
```

```python
print(X[0].toarray()[0].shape)
print((means_resume.shape))
```

```python
from scipy.spatial import distance
for i in range(10):
  print(distance.cosine(X[i].toarray()[0], means_resume))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.spatial import distance
import pandas as pd



num_samples = X.shape[0]
cluster_ids = np.random.randint(10, size=num_samples)

uniqueness_scores = []
for i in range(num_samples):
    current_resume_vector = X[i].toarray()[0]
    # Handle the RuntimeWarning for zero vectors by checking if magnitude is zero
    if np.linalg.norm(current_resume_vector) == 0:
        uniqueness_scores.append(1.0) # Max distance for empty/zero vector
    else:
        uniqueness_scores.append(distance.cosine(current_resume_vector, means_resume))
uniqueness_scores = np.array(uniqueness_scores)


X_dense = X.toarray()
pca = PCA(n_components=2)
components = pca.fit_transform(X_dense)

mean_resume_pca = pca.transform(means_resume.reshape(1, -1))

# Optional: Explained variance ratio
print(f"Explained variance ratio of components: {pca.explained_variance_ratio_}")
print(f"Total explained variance: {pca.explained_variance_ratio_.sum():.2f}")

# --- Prepare Data for Plotting ---
df_plot = pd.DataFrame({
    'PC1': components[:, 0],
    'PC2': components[:, 1],
    'Uniqueness': uniqueness_scores,
    'Type': 'Resume'
})

# Corrected creation of df_mean_plot (wrap scalars in lists)
df_mean_plot = pd.DataFrame({
    'PC1': [mean_resume_pca[0, 0]],
    'PC2': [mean_resume_pca[0, 1]],
    'Uniqueness': [0], # Or some other value, it's the center
    'Type': ['Average Resume']
})


# --- Create the Scatter Plot ---
plt.figure(figsize=(12, 8))

# Plot all resume points (single color)
# Removed 'label' keyword here
sns.scatterplot(
    x='PC1',
    y='PC2',
    size='Uniqueness',   # Size points by uniqueness score (larger for more unique)
    sizes=(50, 500),     # Range of sizes for points
    color='skyblue',     # A single color for all resumes
    alpha=0.7,           # Transparency
    data=df_plot
)

# Plot the mean_resume point (distinct color)
sns.scatterplot(
    x='PC1',
    y='PC2',
    color='red',         # Distinct color for the mean
    marker='X',          # Use a distinct marker (e.g., 'X', 'D', '*')
    s=800,               # Larger size for the mean point
    alpha=1.0,           # Fully opaque
    label='Average Resume', # Label for the legend
    data=df_mean_plot
)

# --- Manual entry for 'All Resumes' in the legend ---
# Create an invisible point just to get a label in the legend
plt.scatter([], [], color='skyblue', s=100, alpha=0.7, label='All Resumes')


plt.title('Resume Distribution with Average Resume Highlighted (PCA Reduced)')
plt.xlabel(f'Principal Component 1 ({pca.explained_variance_ratio_[0]*100:.2f}%)')
plt.ylabel(f'Principal Component 2 ({pca.explained_variance_ratio_[1]*100:.2f}%)')
plt.grid(True, linestyle='--', alpha=0.6)

# Combine legends: the one from sns.scatterplot (for 'Uniqueness' size)
# and the manually added ones (for 'Average Resume' and 'All Resumes')
handles, labels = plt.gca().get_legend_handles_labels()
# Filter out any duplicate labels if seaborn creates them, or just use specific ones
# In this case, we've carefully managed labels, so a direct plt.legend() should work.
plt.legend(title='Point Type', loc='upper right')
plt.tight_layout()
plt.show()
```
